```python
df = df.reset_index(drop=True)
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
vader = SentimentIntensityAnalyzer()
for i in range(len(df)):
    sentence = df.loc[i,'clean_text']
    sent_token = nltk.tokenize.sent_tokenize(sentence)
    c_biden = []
    c_trump = []
    for sub_sent in sent_token:
        if 'biden' in sub_sent.lower():
            b_score = vader.polarity_scores(sub_sent).get('compound')
            c_biden.append(b_score)
        if 'trump' in sub_sent.lower():
            t_score = vader.polarity_scores(sub_sent).get('compound')
            c_trump.append(t_score)
    if len(c_biden) == 0:
        df.loc[i,'Biden sentiment'] = 0
    else:
        df.loc[i,'Biden sentiment'] = np.mean(np.array(c_biden))
    if len(c_trump) == 0:
        df.loc[i,'Trump sentiment'] = 0
    else:
        df.loc[i,'Trump sentiment'] = np.mean(np.array(c_trump))
```
