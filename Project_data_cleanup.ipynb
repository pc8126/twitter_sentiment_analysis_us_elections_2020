```python
import nltk
nltk.download('vader_lexicon')

attributes = ['biden', 'trump', 'covid', 'economy', 'jobs', 'race', 'BLM', 'integrity', 'courts', 'russia', 'china', 'health', 'climate', 'crime', 'wildfires', 'abortion']

def sentiment(item, ii):
    if attributes[ii] in item:
        att_index = item.index(attributes[ii])
        lower_bound = max(att_index - 5, 0)
        upper_bound = min(att_index + 5, len(item))
        text = item[lower_bound:upper_bound+1]
        return vader(' '.join(text))


from nltk.sentiment.vader import SentimentIntensityAnalyzer
analyser = SentimentIntensityAnalyzer()

def vader(item):
    return analyser.polarity_scores(item)['compound']
```
